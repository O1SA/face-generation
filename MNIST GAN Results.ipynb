```python
import tensorflow as tf
from tensorflow.python.tools import inspect_checkpoint as chkp
from mnist_gan import MnistGAN
from matplotlib import pyplot
import helper
```

```python
def display(sample):
    fig=pyplot.figure(figsize=(10, 10), dpi= 80)
    images_grid = helper.images_square_grid(sample, 'L')
    pyplot.axis('off')
    pyplot.imshow(images_grid, cmap='gray')
    pyplot.show()
```

```python
n = MnistGAN()
```

```python
# 2 epochs local
samples = n.generate(100, 'output/mnist_gan.ckpt')
display(samples)
```
